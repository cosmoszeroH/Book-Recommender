# Check out the missingness of "description"

Because it is an important information for natural language processing. So, we would drop the rows that don't have value in "description". However, we should consider the its correlation with other attributes. For example, if books missing descriptions tend to belong to specific categories, authors or years, dropping them could skew the dataset.

```python
import numpy as np
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "average_rating", "age_of_book", "missing_description"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

As we can see, the correlations between "description" and other attributes are not very strong so we are safe to drop.

```python
book_missing = books[~(books["description"].isna()) & 
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]

book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values(by="count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing
```

There are some descriptions that have too few words that can make them meaningless. So, we may drop these rows. And we have to see each bin of words to identify the least number of words that can be meaningful.

```python
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(5, 14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(15, 24), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 34), "description"]
```

From this bin, the decription become reasonably meaningful.

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

As above, "subtitle" is mostly missing, so we would aggregate this field into title.

```python
book_missing_25_words["title_and_subtitle"] = np.where(book_missing_25_words["subtitle"].isna(), 
                                                       book_missing_25_words["title"], 
                                                        book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
book_missing_25_words
```

In following steps, we would use vector search for the decription of book that user wants to match the descriptions' vectors in database. By concatenating the ISBN as ID for the matched description, we could find books easier. 

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)

book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(columns=["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("cleaned_booked.csv", index=False)
)
```
